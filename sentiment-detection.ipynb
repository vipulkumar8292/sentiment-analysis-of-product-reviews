REVIEW CATEGORY---BASED ON STAR RATING, 3.5+ --> 1,

```python
x = amazon['review']
y = amazon['review_category']
```

```python
import re
x_new = []
for w in x:
    w = re.sub('[^a-zA-Z ]','',str(w))
    w = w.lower()
    x_new.append(w)
```

```python
x_new
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test = train_test_split(x_new,y,test_size=0.25,random_state=35)
```

```python
from sklearn.feature_extraction.text import CountVectorizer
```

```python
bags = CountVectorizer(stop_words='english').fit(x_train)
```

```python
x_train_vector = bags.transform(x_train)
```

```python
#x_train_vector.toarray().shape memory error
```

# Implementing Naive Bayes' Model

```python
from sklearn.naive_bayes import MultinomialNB
```

```python
naive = MultinomialNB()
```

```python
naive.fit(x_train_vector,y_train)
```

naive bayes is a fast algorithm

```python
prediction_NB = naive.predict(bags.transform(x_test)) 
```

storing predictions-passing x_test in bag of words


```python
from sklearn.metrics import confusion_matrix,classification_report #importing confusion matrix
```

```python
print(confusion_matrix(prediction_NB,y_test))
```

```python
print(classification_report(prediction_NB,y_test)) #error report
```

r2 is evaluated in regression

```python
review ='good product'
```

```python
print(naive.predict(bags.transform([review])))
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer #importing if-idf
```

```python
tfidf = TfidfVectorizer(min_df=6).fit(x_train) #minimum document frequency=6, rare word is neglected
```

```python
len(tfidf.get_feature_names())
```

# Implementing logistic regression

```python
from sklearn.linear_model import LogisticRegression
```

```python
logis = LogisticRegression()
```

```python
x_train_vector = tfidf.transform(x_train) #converting into tf-idf vector
```

```python
logis.fit(x_train_vector,y_train)
```

```python
prediction_logis = logis.predict(tfidf.transform(x_test)) 
```

```python
print(confusion_matrix(prediction_logis,y_test))
```

```python
print(classification_report(prediction_logis,y_test)) #error report
```

```python
review2 = 'very nice product'
```

```python
print(logis.predict(tfidf.transform([review2])))
```
